```python
class pixel :

    def __init__ (self):
        self._x = 0
        self._y = 0
        self._red = 0
        self._green = 0
        self._blue = 0

    def set_coords(self, x, y):
        self._x = x
        self._y = y
    
    def pixel(self,red,green,blue) :
        self._red = red
        self._green = green
        self.blue = blue

    def set_grayscale(self):
        grey = (self._red + self._blue + self._green)//3
        self._red = grey 
        self._green = grey 
        self._blue = grey

    def print_pixel_info(self):
        # color = ''.join(['red' if  self._blue == 0 and self._green == 0 and self._red > 50 'blue' if self._red == 0 and self._green == 0 and self._blue > 50 'green' if self.red == 0 and self._blue == 0 and self._green > 50 ])
        # print()
        if self._red == 0 and self._blue == 0 and self._green > 50 : 
            print('x:', self._x, 'y:', self._y, 'color:', (self._red,self._green,self._blue) ,'Green')
        if self._red == 0 and self._green == 0 and self._blue > 50 :
            print('x:', self._x, 'y:', self._y, 'color:', (self._red,self._green,self._blue) ,'Blue')
        if self._green == 0 and self._blue == 0 and self._red > 50 :
            print('x:', self._x, 'y:', self._y, 'color:', (self._red,self._green,self._blue) ,'Red')
        else :
            print('x:', self._x, 'y:', self._y,'color:', (self._red,self._green,self._blue))

def main() :
    p = pixel()
    p.set_coords(5,6)
    p.pixel(250,0,0)
    p.print_pixel_info()
    p.set_grayscale()
    p.print_pixel_info()
main()
```

```python
class BigThing :
    def __init__(self,var) :
        self._var = var
    def size(self) :
        if isinstance(self._var,int) == True :
            return (self._var)
        else :
            return(len(self._var))

class BigCat(BigThing) :
    def __init__(self,var,weight) :
        BigThing.__init__(self,var)
        self._weight = weight
    def size (self):
        if self._weight > 15 and self._weight <= 20:
            return 'Fat'
        elif self._weight > 20 :
            return 'Very Fat' 
        else :
            return 'OK'


my_thing = BigThing("balloon")
print(my_thing.size())
cutie = BigCat("mitzy", 1)
print(cutie.size())
```

```python
class Animal :
    """ A class used to represent an animal """ 

    zoo_name = "Hayaton"

    def __init__ (self,_name,_hunger=0):
        """ initialize the name of the animal and the degree of hunger
        param _name  : the name of the animal
        param _hunger :the degree of animal's hunger (default = 0)
        type _name : str
        type _hunger : int
        return : None 
        rtype : None 
        """
        self._hunger = _hunger
        self._name = _name

    def get_name(self) :
        """ return the name of the animal 
        return : _name
        rtype : str 
        """
        return self._name

    def is_hungry(self) :
        """ return boolean value which describes if the animal is hungry or not
        return : True / False
        rtype : Boolian value
        """
        if self._hunger > 0 :
            return True
        else :
            return False
    
    def feed(self) :
        """ lower degree of hungr one point
        return : None
        rtype : None 
        """
        self._hunger -= 1

    def talk(self) :
         """ doesn't do anything
         return : None
         rtype : None
         """
         pass

class Dog(Animal) :
    """ A class used to represent a dog """
    def __init__(self,_name,_hunger):
        """ inherit the initialize method of the super class
        param _name : the name of the animal
        param _hunger : the degree of animal's hunger (default = 0)
        type _name : str
        type _hunger : int
        return : None
        rtype : None
        """
        Animal.__init__(self,_name,_hunger)

    def talk(self) :
        """ inherit the talk method of the super class and print woof woof
        return : woof woof
        rtype : str 
        """
        Animal.talk(self)
        print('woof woof')

    def fetch_stick(self) : 
        """ print There you go, sir!
        return : There you go, sir!
        rtype : str
        """
        print('There you go, sir!')

class Cat(Animal) :
    """ A class used to represent a cat """
    def __init__(self,_name,_hunger):
        """ inherit the initialize method of the super class
        param _name : the name of the animal
        param _hunger : the degree of animal's hunger (default = 0)
        type _name : str
        type _hunger : int
        return : None
        rtype : None
        """
        super().__init__(_name,_hunger)

    def talk(self) :
        """ inherit the talk method of the super class and print meow
        return : meow
        rtype : str 
        """
        Animal.talk(self)
        print('meow')

    def chase_laser(self) :
        """ print meeeeow
        return : meeeeow
        rtype : str
        """
        print('Meeeeow')   

class Skunk(Animal) :
    """ A class used to represent a skunk """
    def __init__(self,_name,_hunger,_stink_count=6):
        """ inherit the initialize method of the super class and initialize the skunk's stinks count 
        param _name : the name of the animal
        param _hunger : the degree of animal's hunger (default = 0)
        param _stink_count : the number of stinks (defalut = 6)
        type _name : str
        type _hunger : int
        type _stink_count : int
        return : None
        rtype : None
        """
        Animal.__init__(self,_name,_hunger)
        self._stink_count = _stink_count

    def talk(self) :
        """ inherit the talk method of the super class and print tsssss
        return : tsssss
        rtype : str
        """
        Animal.talk(self)
        print('tsssss')

    def stink(self) :
        """ print Dear lord!
        return : Dear lord!
        rtype : str 
        """
        print('Dear lord!')

class Unicorn(Animal) :
    """ A class used to represent a unicorn """
    def __init__(self,_name,_hunger):
        """ inherit the initialize method of the super class
        param _name : the name of the animal
        param _hunger : the degree of animal's hunger (default = 0)
        type _name : str
        type _hunger : int
        return : None
        rtype : None
        """
        Animal.__init__(self,_name,_hunger)

    def talk(self) :
        """ inherit the talk method of the super class and print Good day, darling
        return : Good day, darling
        rtype : str
        """
        Animal.talk(self)
        print('Good day, darling')

    def sing(self) :
        """ print I'm not your toy... 
        retrun : I'm not your toy...
        rtype : str
        """
        print('I’m not your toy...')

class Dragon(Animal) :
    """ A class used to represent a dragon """
    def __init__(self,_name,_hunger,_color='Green'):
        """ inherit the initialize method of the super class and initialize the dragon's color
        param _name : the name of the animal
        param _hunger : the degree of animal's hunger (default = 0)
        param _color : the color of the dragon (defalt = green)
        type _name : str
        type _hunger : int
        type _color : str
        return : None
        rtype : None
        """
        Animal.__init__(self,_name,_hunger)
        self._color = _color

    def talk(self) :
        """ inherit the talk method of the super class and print Raaaawr 
        return : Raaaawr
        rtype : str
        """
        Animal.talk(self)
        print('Raaaawr')

    def breath_fire (self) :
        """ print $@#$#@$
        return : $@#$#@$
        rtype : str
        """
        print('$@#$#@$')

def main() :
    """ create animals and initialize them. save them in list. if the animal is hungry     it prints the animal's type and name . it feeds the animal until it's not hungry.
    it prints the words the animal knows how to say. call to the animal's special          method. print the name of the zoo. 
    this function uses in the is_hungry() method, get_name() method, feed() method,        talk() method, fetch_stick() method, chase_laser() method, stink() method, sing()      method and breath_fire() method

    return : animal's type and name, the words the animal knows how to say, the animal's special method ans it prints the zoo name.
    rtype : str
    """
    Brownie = Dog('Brownie',10)
    Zelda = Cat('Zelda',3)
    Stinky = Skunk('Stinky',0)
    Keith = Unicorn('Keith',7)
    Lizzy = Dragon('Lizzy',1450)
    Doggo = Dog('Doggo',80)
    Kitty = Cat('Kitty',80)
    StinkyJr = Skunk('Stinky Jr.',80)
    Clair = Unicorn('Clair',80)
    McFly = Dragon('McFly',80)

    zoo_lst = []

    zoo_lst.append(Brownie)
    zoo_lst.append(Zelda)
    zoo_lst.append(Stinky)
    zoo_lst.append(Keith)
    zoo_lst.append(Lizzy)
    zoo_lst.append(Doggo)
    zoo_lst.append(Kitty)
    zoo_lst.append(StinkyJr)
    zoo_lst.append(Clair)
    zoo_lst.append(McFly)
    
    for animal in zoo_lst :
        if animal.is_hungry() :
            print(type(animal).__name__,animal.get_name())
        while animal.is_hungry() :
            animal.feed()
        animal.talk()
        if isinstance(animal,Dog) :
            animal.fetch_stick()
        if isinstance(animal,Cat) :
            animal.chase_laser()
        if isinstance(animal,Skunk) :
            animal.stink()
        if isinstance(animal,Unicorn) :
            animal.sing()
        if isinstance(animal,Dragon) :
            animal.breath_fire()
    print(Animal.zoo_name)
if __name__ == '__main__' :
    main()
```
